```python
import cv2
import os
import math
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

file_name = './material/House.png'

# Load a grayscale image
image = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)
image_file = Image.open(file_name)
orig_image_size = str(len(image_file.fp.read()))

# Display the original image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image Size : ' + orig_image_size + ' bytes')
plt.axis('off')

# Bit-plane slicing: Removing the first two bits
# As we've seen in histogram, masking up to 3-bit value intensities may not affect the image very much

bit_mask = 0b11111000  # This keeps the top 5 bits and sets the three least significant bits (LSBs) to 0
compressed_image = cv2.bitwise_and(image, bit_mask)
cv2.imwrite('House-compressed.png' ,compressed_image)

image_sec_file = Image.open('House-compressed.png')
sec_image_size = str(len(image_sec_file.fp.read()))

# Display the compressed image (first two bits removed)
plt.subplot(1, 2, 2)
plt.imshow(compressed_image, cmap='gray')
plt.title('Image with First Two Bits Removed Size : ' + sec_image_size + ' bytes')
plt.axis('off')

plt.show()
```

 # Task 2 - Histogram equalization

```python
# Read the image in grayscale
low_contrast_image = cv2.imread('./material/cant-see-me.png', cv2.IMREAD_GRAYSCALE)


# Step 3: Show the original low-contrast image and its histogram
def plot_histogram(image, title):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    plt.plot(hist)
    plt.xlim([0, 256])
    plt.title(title)
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

# Display the original image and histogram
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(low_contrast_image, cmap='gray')
plt.title('Original Low-Contrast Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plot_histogram(low_contrast_image, 'Histogram of Original Image')
plt.show()

# Step 4: Apply Histogram Equalization
equalized_image = cv2.equalizeHist(low_contrast_image)
cv2.imwrite("can-see-me.png",equalized_image)

# Step 5: Show the equalized image and its histogram
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plot_histogram(equalized_image, 'Histogram of Equalized Image')
plt.show()
```
